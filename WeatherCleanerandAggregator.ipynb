## This file cleans the weather data to make them universal, as some months had data that was unique to itself. It also aggregates the individual months into one directory.

```python
import pandas as pd
```

```python
path = "~/Documents/MSCA/3spring2019/BigData/Final_Project/old_weather/"
```

```python
november = pd.read_csv(path+"november_weather.csv", delimiter = ",",header=0)
december = pd.read_csv(path+"december_weather.csv", delimiter = ",",header=0)
january = pd.read_csv(path+"january_weather.csv", delimiter = ",",header=0)
february = pd.read_csv(path+"february_weather.csv", delimiter = ",",header=0)
march = pd.read_csv(path+"march_weatherUpdated.csv", delimiter = ",",header=0)
```

```python
len(november)+len(december)+len(january)+len(february)+len(march)
```

```python
december.windBearing = december.windBearing.astype("float64")
```

```python
months = [november, december, january, february, march]
result = pd.concat(months)
result = result.sort_values("time")
len(result)
```

```python
result["Unnamed: 0"] = range(0,len(result))
```

#### Imputing missing values for the enitre five months on five features.

```python
listprob = result["Unnamed: 0"][result.precipProbability.isnull()]
for i in listprob:
    j = i + 1
    k = i - 1
    result.iloc[i,8] = (result.iloc[k,8] + result.iloc[j,8])/2
listint = result["Unnamed: 0"][result.precipIntensity.isnull()]
for i in listint:
    j = i + 1
    k = i - 1
    result.iloc[i,7] = (result.iloc[k,7] + result.iloc[j,7])/2 
listtemp = result["Unnamed: 0"][result.temperature.isnull()]
for i in listtemp:
    j = i + 1
    k = i - 1
    result.iloc[i,12] = (result.iloc[k,12] + result.iloc[j,12])/2
listhumid = result["Unnamed: 0"][result.humidity.isnull()]
for i in listhumid:
    j = i + 1
    k = i - 1
    result.iloc[i,4] = (result.iloc[k,4] + result.iloc[j,4])/2 
listAtemp = result["Unnamed: 0"][result.apparentTemperature.isnull()]
for i in listAtemp:
    j = i + 1
    k = i - 1
    result.iloc[i,1] = (result.iloc[k,1] + result.iloc[j,1])/2 
```

#### Checking for missing values that will be needed for modeling, otherwise filling them with zero's to be dropped later

```python
total = result.isnull().sum().sort_values(ascending=False)
percent_1 = result.isnull().sum()/result.isnull().count()*100
percent_2 = (round(percent_1, 1)).sort_values(ascending=False)
missing_data = pd.concat([total, percent_2], axis=1, keys=['Total', '%'])
missing_data.head(15)
```
